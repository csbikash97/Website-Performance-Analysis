```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("data-export (1).csv")
```

```python
df.columns = df.iloc[0]
df.head(10)
```

```python
df = df.drop(index = 0).reset_index(drop = True)
df.columns= ["channel group","DateHour","Users","Sessions","Engaged Sessions","Average engagement time per session","Engaged sessions per user","Events per session","Engagement rate","Event count"]
```

```python
df.head()
```

```python
df.info()
```

```python
df["DateHour"]=pd.to_datetime(df["DateHour"],format = "%Y%m%d%H", errors = 'coerce')
```

```python
df.info()
```

```python
df.head()
```

```python
numeric_cols = df.columns.drop(["channel group", "DateHour"])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric,errors = 'coerce')
df["Hour"] = df["DateHour"].dt.hour
```

# Q)- What pattern or trends can you observe in website session and users over time?

# Sessions and users over time 

```python
sns.set(style = "whitegrid")
```

```python
plt.figure(figsize=(8,4))
df.groupby("DateHour")[["Sessions","Users"]].sum().plot(ax=plt.gca())
plt.title("Session and users over time")
plt.xlabel("DateHour")
plt.ylabel("count")
plt.show()
```

# Q)- which marketing channel brought the highest number of users to the websites, and how can we use this insights to         improve traffic from other source?

# Total users by channels

```python
plt.figure(figsize = (6,4))
sns.barplot(data = df,x="channel group", y="Users",estimator=np.sum,palette ="viridis")
plt.title("Total Users By channels")
plt.xticks(rotation = 45)
```

# Q)- Which channel has the highest Average engagement time, and what does that tell usabout user behavior and                 effectiveness?

# Average angement time by channel

```python
df.head()
```

```python
plt.figure(figsize= (8,5))
sns.barplot(data = df,x="channel group",y = "Average engagement time per session",estimator= np.mean, palette ="magma")
plt.title("Average engagement Time by channel")
plt.xticks(rotation = 45)
plt.show()
```

# Q)- How does engagement rate vary across different Traffic channels?

# Engedgement Rate Distribution by channel

```python
plt.figure(figsize= (8,5))
sns.boxplot(data = df, x= "channel group",y = "Engagement rate",palette = "coolwarm")
plt.title("Engagement Rate Distribution by channel")
plt.xticks(rotation = 40)
plt.show()
```

# Q)- which channels are driving more engaged sessions compared to non-engaged ones, and what strategies can improve           engagement in underperforming channels?

# Engaged vs non engaged sessions


```python
session_df = df.groupby("channel group")[["Sessions","Engaged Sessions"]].sum().reset_index()
session_df["Non-Engaged"]   = session_df["Sessions"]-session_df["Engaged Sessions"]
session_df_melted = session_df.melt(id_vars="channel group",value_vars=["Engaged Sessions","Non-Engaged"])

plt.figure(figsize=(8, 5))
sns.barplot(data = session_df_melted,x = "channel group",y = "value",hue= "variable")
plt.title("engage vs non engaged sessions")
plt.xticks(rotation = 45)
plt.show()
```

# Q)- At what hours of the day does each channels drive the most traffic?

# Traffic by hour and channel

```python
heatmap_date = df.groupby(["Hour","channel group"])["Sessions"].sum().unstack().fillna(0)
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_date,cmap="YlGnBu",linewidths =.5,annot = True, fmt = '.0f')
plt.title("Traffic by hour and channel")
plt.xlabel("channel group")
plt.ylabel("Hour of Day")
plt.show()
```

# Q)- is there any correlation between high traffic(session) and high engagement rate over time?

# Engagement rate vs sessions over time

```python
df_plot = df.groupby("DateHour")[["Engagement rate","Sessions"]].mean().reset_index()
plt.figure(figsize=(10,5))
plt.plot(df_plot["DateHour"],df_plot["Engagement rate"],label= "Engagement rate", color= "blue")
plt.plot(df_plot["DateHour"],df_plot["Sessions"],label= "Sessions", color= "green")
plt.title("Engagement rate vs sessions over time")
plt.xlabel("DateHour")
plt.legend()
plt.grid(True)
plt.show()
```
